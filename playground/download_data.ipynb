```python
# Create a table that contains the api of each station for each feature (row: station, column: feature), and save it as a csv file

with open(f'{dict_dir}/code_to_station.json', 'r') as f:
    code_to_station = json.load(f)
all_stations = list(code_to_station.values())

with open(f'{dict_dir}/all_years_links.json', 'r') as f:
    all_years_links = json.load(f)

table = pd.DataFrame(columns=all_years_links, index=all_stations)
print(f'initial table: {table}')

for file in os.listdir(apis_dir):
    if file.endswith('.csv'):
        df = pd.read_csv(f'{apis_dir}/{file}')
        columns = df.columns
        useful_columns = [col for col in columns if col in all_years_links]
        for index, row in df.iterrows():
            for col in useful_columns:
                table.loc[row['WeatherStationName_en'], col] = row[col]

print(f'final table: {table}')

# save the table to a csv file
table.to_csv(f'{dict_dir}/station_feature_to_api_table.csv', index=True)


```
